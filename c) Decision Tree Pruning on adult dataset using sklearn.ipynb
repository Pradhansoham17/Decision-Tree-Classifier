```python
# after removing data row as index row and changing column names
import pandas as pd
import numpy as np
df=pd.read_csv('C:\\Users\\soham\\Downloads\\adult.data',names=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','income'],header=None)
df  
```

```python
#replacing "?" by np.nan
df=df.replace(' ?',np.nan)
#replace missing values by mean,mode
for i in df.columns:
    if (type(df[i].dropna().tolist()[0])==str): #we are checking type of first value of each column after dropping NA's from it.
        df[i]=df[i].fillna(df[i].mode()[0])   #So, no problem even if first value of column turns out to be missing
    elif (type(df[i].dropna().tolist()[0])==float):  #we are checking type of first available element of that column using dropna 
        df[i]=df[i].fillna(df[i].mean())
df
```

```python
#creating a dataframe with only integer/numerical type columns
dfint=pd.DataFrame()
for i in df.columns:
    if (type(df[i].tolist()[0])==int):
        dfint[i]=df[i].tolist()
dfint
```

```python
#creating a dataframe with only integer/numerical type columns
dfobj=pd.DataFrame()
for i in df.columns:
    if (type(df[i].tolist()[0])==str):
        dfobj[i]=df[i].tolist()
dfobj
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(handle_unknown='ignore')
X = dfobj.iloc[:,:-1]
enc.fit(X)
```

```python
#converting enc.transform array into a proper dataframe
onehotlabels = enc.transform(X).toarray()

new_columns=[]
for col, values in zip(X.columns, enc.categories_):
    new_columns.extend([col + '_' + str(i) for i in values])

    
df_ohe= pd.concat([dfint,pd.DataFrame(onehotlabels, columns=new_columns)], axis='columns')
df_ohe= df_ohe.rename(columns={"native-country_ Trinadad&Tobago": "native-country_ Trinadad-Tobago"})
df_ohe
```

```python
df_comb_ohe = pd.concat([df_ohe, df.iloc[:,[-1]]], axis=1)
df_comb_ohe
```

<b><u><font size='7'>Various Pruning Strategies :- </font></u></b>

<b><font size='5'>1) max_depth :-</font></b>

<b><font size='3'>The maximum depth of the tree. </font></b>

```python
from sklearn.tree import DecisionTreeClassifier
clf1 = DecisionTreeClassifier(criterion="entropy",max_depth=3,random_state=0)
X1=df_ohe.iloc[:,:]
y1=df.iloc[:,[-1]]
clf1.fit(X1,y1)
```

```python
from sklearn import tree
clf1=clf1.fit(X1,y1)
tree.plot_tree(clf1) 
```

```python
clf1.tree_.max_depth
```

```python
#https://towardsdatascience.com/scikit-learn-decision-trees-explained-803f3812290d
#check above website to get proper explaination of graphviz trees
import graphviz 
dot_data = tree.export_graphviz(clf1,feature_names=df_ohe.columns,  
                                class_names=['0','1'],filled=True, rounded=True,  
                                special_characters=True, out_file=None) 
#dont give class names as clf.classes_ or (<=50K, >50K), give (0,1) instead
graph = graphviz.Source(dot_data) 
graph.render("adult_max_depth") 
```

<b><font size='5'>2) min_samples_split :-</font></b>

<b><font size='3'>The minimum number of samples required to split an internal node:

If int, then consider min_samples_split as the minimum number.

If float, then min_samples_split is a fraction and ceil(min_samples_split * n_samples) are the minimum number of samples for each split.</font></b>

```python
from sklearn.tree import DecisionTreeClassifier
clf2 = DecisionTreeClassifier(criterion="entropy",min_samples_split=5000,random_state=0)
X2=df_ohe.iloc[:,:]
y2=df.iloc[:,[-1]]
clf2.fit(X2,y2)
```

```python
from sklearn import tree
clf2=clf2.fit(X2,y2)
tree.plot_tree(clf2) 
```

```python
#https://towardsdatascience.com/scikit-learn-decision-trees-explained-803f3822290d
#check above website to get proper explaination of graphviz trees
import graphviz 
dot_data = tree.export_graphviz(clf2,feature_names=df_ohe.columns,  
                                class_names=['0','1'],filled=True, rounded=True,  
                                special_characters=True, out_file=None) 
#dont give class names as clf.classes_ or (<=50K, >50K), give (0,1) instead
graph = graphviz.Source(dot_data) 
graph.render("adult_min_samples_split") 
```

<b><font size='5'>3) min_samples_leaf :-</font></b>

<b><font size='3'>The minimum number of samples required to be at a leaf node. 

If int, then consider min_samples_leaf as the minimum number.

If float, then min_samples_leaf is a fraction and ceil(min_samples_leaf * n_samples) are the minimum number of samples for each node.</font></b>

```python
from sklearn.tree import DecisionTreeClassifier
clf3 = DecisionTreeClassifier(criterion="entropy",min_samples_leaf=1000,random_state=0)
X3=df_ohe.iloc[:,:]
y3=df.iloc[:,[-1]]
clf3.fit(X3,y3)
```

```python
from sklearn import tree
clf3=clf3.fit(X3,y3)
tree.plot_tree(clf3) 
```

```python
#https://towardsdatascience.com/scikit-learn-decision-trees-explained-803f3822290d
#check above website to get proper explaination of graphviz trees
import graphviz 
dot_data = tree.export_graphviz(clf3,feature_names=df_ohe.columns,  
                                class_names=['0','1'],filled=True, rounded=True,  
                                special_characters=True, out_file=None) 
#dont give class names as clf.classes_ or (<=50K, >50K), give (0,1) instead
graph = graphviz.Source(dot_data) 
graph.render("adult_min_samples_leaf") 
```

<b><font size='5'>4) min_impurity_decrease :-</font></b>

<b><font size='3'>A node will be split if this split induces a decrease of the impurity greater than or equal to this value. </font></b>

```python
from sklearn.tree import DecisionTreeClassifier
clf4 = DecisionTreeClassifier(criterion="entropy",min_impurity_decrease=0.000105,random_state=0)
X4=df_ohe.iloc[:,:]
y4=df.iloc[:,[-1]]
clf4.fit(X4,y4)
```

```python
from sklearn import tree
clf4=clf4.fit(X4,y4)
tree.plot_tree(clf4) 
```

```python
#https://towardsdatascience.com/scikit-learn-decision-trees-explained-803f3822290d
#check above website to get proper explaination of graphviz trees
import graphviz 
dot_data = tree.export_graphviz(clf4,feature_names=df_ohe.columns,  
                                class_names=['0','1'],filled=True, rounded=True,  
                                special_characters=True, out_file=None) 
#dont give class names as clf.classes_ or (<=50K, >50K), give (0,1) instead
graph = graphviz.Source(dot_data) 
graph.render("adult_min_impurity_decrease") 
```
